#1. What is a parameter?
  - A parameter is a fixed value that is used in model to describe a specific characteristic of a dataset or system. In data analytics, parameters helps to define the structure of a statistical or machine-learning model. For example, in a linear regression equation (y = mx + b), m (slope) and b (intercept) are parameters.

#2. What is correlation? What does negative correlation mean?
  - Correlation measures the relationship between two variables—how they change together. If one of the variable increases while the other also increases, they are positively correlated. If one increases while the other decreases, they are negatively correlated. For example:
   * Positive correlation → More hours studied → Higher exam scores
   * Negative correlation → More time on social media → Lower exam scores


  - A negative correlation means that as one variable increases, the other decreases.For example:
    * The more you exercise, the less body fat you have.
    * The more discounts a company gives, the lower its profits.

The strength of a negative correlation is measured between -1 to 0 in a correlation coefficient.

#3. Define Machine Learning. What are the main components in Machine Learning?
  - Machine Learning (ML) is a field of artificial intelligence that allows computers to learn from data and make decisions without being explicitly programmed.
  - Main Components of Machine Learning:
    * Data-The foundation of ML; the more relevant the data, the better the model.
    * Algorithms-The set of rules that the model follows to find patterns.
    * Model-The system that learns from data and makes predictions.
    * Training-The process of feeding data into the model to help it learn.
    * Loss Function-measure of how well (or badly) the model is performing.
    * Optimization-Adjusting parameters to improve accuracy.
#4. How does loss value help in determining whether the model is good or not?
  - The loss value tells us how far off the model’s predictions are from the actual values.
    * High loss → Model is making many mistakes (bad model).
    * Low loss → Model is making fewer mistakes (good model).
    
A good model has a low loss value, meaning its predictions are close to reality.

#5. What are continuous and categorical variables?
  - Continuous Variables: Variables that can take any numeric value within a range. For example: Height (170.5 cm), Weight (65.2 kg), Temperature (36.6°C)
  - Categorical Variables: Variables that represent categories or groups. For example: Gender (Male, Female, Other), Colors (Red, Blue, Green), Product Categories (Electronics, Clothing, Grocery)

#6. How do we handle categorical variables in Machine Learning? What are the common techniques?
  - Categorical variables contain text-based data (e.g., "Male", "Female", "Red", "Blue"). Machine Learning models cannot process text, so we need to convert categorical variables into numbers.
  - Common techniques to handle categorical variables:
    * Label Encoding → Assigns a unique number to each category.For example: {Red → 0, Blue → 1, Green → 2}
    * One-Hot Encoding → Creates separate columns for each category and marks them with 0 or 1. For example:
                Color	Red	Blue	Green
                Red	   1   0    	0
                Blue	 0	 1    	0
                Green	 0 	 0    	1
    * Ordinal Encoding → Used for ranked categories (e.g., Small < Medium < Large → {Small = 1, Medium = 2, Large = 3}).

#7. What do you mean by training and testing a dataset?
  - Training dataset: The data used to teach the model patterns and relationships.
  - Testing dataset: The data used to check if the model learned correctly by making predictions on unseen data.

Think of it like preparing for an exam:
  - Studying (Training data) → Learning concepts.
  - Taking the exam (Testing data) → Checking if you can apply what you learned.

#15. What is causation? Explain difference between correlation and causation with an example.
   - Causation means that one event directly causes another event to happen.
   - Correlation means that two variables are related but one does not necessarily cause the other.
   - Example:
      1. Correlation: Ice cream sales and drowning cases increase in summer.
           * (Does ice cream cause drowning? No! It is just that both happen more in hot weather.)
      2. Causation: Smoking leads to lung cancer.
           * (Smoking directly damages lungs and increases cancer risk.)
   - Key Difference: Correlation shows a relationship, while causation proves direct influence.

#16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
   - An optimizer is a function that adjusts the parameters of a Machine Learning model to minimize errors (loss function) and improve accuracy.
   - Types of Optimizers:
     1. Gradient Descent- Adjusts weights by finding the direction that reduces error the fastest.
        * Example: A ball rolling down a hill to reach the lowest point.
     2. Stochastic Gradient Descent (SGD)- Updates model weights using one data point at a time.
     3. Adam (Adaptive Moment Estimation)- A faster, adaptive method that adjusts learning rates dynamically.
        * Example: Used in deep learning models like neural networks.

#21. What is feature scaling? How does it help in Machine Learning?
   -  Feature Scaling standardizes data so that all variables have the same scale. It prevents large values from dominating small ones.
   - Example:
        * Suppose age is [20, 30, 40] and income is [30,000, 50,000, 100,000].
        * Since income values are much larger, the model may give more importance to income than age.
        * Scaling fixes this issue by converting both to the same range.
